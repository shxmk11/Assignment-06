# Binary Search Tree 연습 문제

01 다음 이진 검색 트리에 대해 답하시오.

1) 이 이진 검색 트리의 리프 노드는 총 몇개인가?

➡️ 리프 노드: 자식이 없는 노드

➡️ 7, 17, 70

▶️ 3개

2) 현재 높이 4를 유지하면서 키를 추가한다면 가능한 리프 노드의 최대 개수는 몇 개인가?

➡️ 포화 이진 트리가 되었을 때가 가능한 리프 노드의 최대 개수가 나옴

▶️ 2의 3승 = 8

3) 높이 4인 이진 검색 트리가 포화 이진 트리 모양이 되면 리프 노드와 리프 노드를 제외한 나머지 노드(내부 노드라 한다)의 수는 몇 개나 차이가 나는가?

➡️ 리프 노드의 수: 2의 3승 = 8

➡️ 내부 노드의 수: 전체 노드의 수 - 리프 노드의 수 = (2의 4승 - 1) - 8 = 15 - 8 = 7

▶️ 리프 노드의 수 - 내부 노드의 수 = 8 - 7 = 1

4) 위 (3)의 성질은 높이가 ℎ인 모든 포화 이진 트리에 대해 성립하는가?

➡️ 모든 포화 이진 트리에서 리프 노드의 수는 [2의 (h - 1)승]이고, 내부 노드의 수는 [2의 (h - 1)승 - 1]이기 때문에 항상 성립.

▶️ O.


02 임의의 완전 이진 트리에 대한 다음 물음에 답하시오.

1) 트리의 총 노드 수가 𝑛이면 리프 노드의 수는 몇 개인가?

▶️ n/2

2) 위 (1)에서 리프 노드의 수와 내부 노드의 수는 어떤 관계인가?

 n이 짝수일 때, 리프 노드 수 = 내부 노드 수

▶️ n이 홀수일 때, 리프 노드 수 = 내부 노드 수 + 1


03 다음 이진 검색 트리에 대해 답하시오.


1) 이 검색 트리를 전위 순회 방식으로 순회할 때 노드의 방문 순서를 나열하시오.
순서는 노드에 있는 킷값으로 적으시오.

➡️ 전위 순회 방식: 루트 노드 → 왼쪽 → 오른쪽

▶️ 방문 순서: 5, 3, 3, 8, 7, 6, 9

2) 이 검색 트리를 중위 순회 방식으로 순회할 때 노드의 방문 순서를 나열하시오.

➡️ 중위 순회 방식: 왼쪽 → 루트 노드 → 오른쪽

▶️ 방문 순서: 3, 3, 5, 6, 7, 8, 9

3) 이 검색 트리를 후위 순회 방식으로 순회할 때 노드의 방문 순서를 나열하시오.

➡️ 후위 순회 방식: 왼쪽 → 오른쪽 → 루트 노드

▶️ 방문 순서: 3, 3, 6, 7, 9, 8, 5

4) 중위 순회는 모든 이진 검색 트리에 대해 항상 키를 정렬된 순서로 방문한다는
사실을 증명하시오.

➡️ 왼쪽 서브 트리의 모든 키는 루트 노드보다 작고, 오른쪽 서브 트리의 모든 키는 루트 노드보다 크다. 중위 순회는 왼쪽 → 루트 노드 → 오른쪽 순서로 진행되기 때문에 항상 정렬된 순서로 키를 방문하게 된다.




04 다음 이진 검색 트리에서 키 30을 검색하는 과정을 순서대로 설명하시오.

➡️ 노드 15와 비교: 30 > 15 → 오른쪽 자식으로 이동

➡️ 노드 28과 비교: 30 > 28 → 오른쪽 자식으로 이동

▶️ 노드 30 발견

08 6번 문제에서 주어진 이진 검색 트리에서 키 15를 삭제한 후의 모양을 그리시오.

➡️ 키 15의 직후 원소 = 28 → 키 15 삭제 → 키 28을 키 15 자리로 옮김 → 키 28이 있던 자리에 키 28의 자식을 놓음



09 𝑛개의 원소로 구성되는 이진 검색 트리의 가능한 최대 높이와 최소 높이는 어떻게
되는가?

▶️ 한쪽으로 쏠린 경우 → 최대 높이: n

▶️ 완전 이진 트리인 경우 → 최소 높이: log₂(n+1)
